```python
miles = int(input())
```

```python
print("\nEnter miles per hour: ")
```

```python
mph = int(input())
```

```python
print("\n\nEstimated Travel Time")
```

```python
total_hours = float(miles / mph)
```

```python
hours = int(total_hours)
```

```python
print("\nHours: ")
```

```python
print(hours)
```

```python
minutes = round((total_hours - hours) * 60)
```

```python
print("\nMinutes: ")
```

```python
print((int(minutes)))
```
