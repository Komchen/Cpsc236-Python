```python
print("Hours Worked:\t")
```

```python
hours_worked = int(input())
```

```python
print("\nHourly Pay Rate:\t")
```

```python
pay_rate = float(input())
```

```python
print("\n\nGross Pay: \t")
```

```python
gross_pay = float(hours_worked * pay_rate)
```

```python
print(gross_pay)
```

```python
print("\nTax Rate:\t")
```

```python
tax_rate = 18
```

```python
print(str(tax_rate) + "%")
```

```python
print("\nTax Amount:\t")
```

```python
tax_amt = gross_pay * float(tax_rate / 100)
```

```python
print(tax_amt)
```

```python
print("\nTake Home Pay:\t")
```

```python
home_pay = gross_pay - tax_amt
```

```python
print("$" + str(home_pay))
```
